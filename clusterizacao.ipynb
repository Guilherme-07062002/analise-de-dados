```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
```

```python
# Carregar o dataset wine
wine = load_wine()
X_wine = wine.data
```

```python
# Lista para armazenar os scores
silhouette_scores = []
davies_bouldin_scores = []

# Loop para testar diferentes números de clusters
for n_clusters in range(2, 7):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_wine)
    labels = kmeans.labels_

    # Calcular as métricas
    silhouette = silhouette_score(X_wine, labels)
    davies_bouldin = davies_bouldin_score(X_wine, labels)

    silhouette_scores.append(silhouette)
    davies_bouldin_scores.append(davies_bouldin)

# Exibir os resultados
print("Silhouette Scores:", silhouette_scores)
print("Davies-Bouldin Scores:", davies_bouldin_scores)
```

```python
# Plotando os resultados
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(range(2, 7), silhouette_scores, marker='o')
plt.title("Silhouette Score")
plt.xlabel("Número de Clusters")
plt.ylabel("Score")

plt.subplot(1, 2, 2)
plt.plot(range(2, 7), davies_bouldin_scores, marker='o')
plt.title("Davies-Bouldin Score")
plt.xlabel("Número de Clusters")
plt.ylabel("Score")

plt.tight_layout()
plt.show()
```
