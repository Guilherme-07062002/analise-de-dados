### Lidando com valores importantes

Para levar em consideração quais valores são mais importantes, foi lido a [Documentação](https://archive.ics.uci.edu/dataset/544/estimation+of+obesity+levels+based+on+eating+habits+and+physical+condition) a fim de verificar se algum deles vai precisar sair antes dos próximos passos.

#### Procurando por "outliers"

```python
df["Gender"] = df["Gender"].replace({"Male": 0, "Female": 1})
```

```python
df["Age"].plot(kind="box")
```

```python
df = df[df["Age"] < df["Age"].quantile(0.95)]
```

```python
df["Weight"].plot(kind="box")
```

```python
df["Height"].plot(kind="box")
```

#### Usando enums para transformar strings em inteiros

```python
print("Mostrando todos os valores disponíveis do CAEC: ", df['CAEC'].unique())
```

```python
df["CAEC"] = df["CAEC"].replace({"no": 0, "Sometimes": 1, "Frequently": 2, "Always": 3})
```

```python
print("Mostrando todos os valores disponíveis do CALC: ", df['CALC'].unique())
```

```python
df["CALC"] = df["CALC"].replace({"no": 0, "Sometimes": 1, "Frequently": 2, "Always": 3})
```

```python
print("Mostrando todos os valores disponíveis do NObeyesdad: ", df['NObeyesdad'].unique())
```

```python
df["NObeyesdad"] = df["NObeyesdad"].replace(
    {
        "Insufficient_Weight": 0, 
        "Normal_Weight": 1, 
        "Overweight_Level_I": 2, 
        "Overweight_Level_II": 3, 
        "Obesity_Type_I": 4, 
        "Obesity_Type_II": 5, 
        "Obesity_Type_III": 6
    })
```

```python
print("Mostrando todos os valores disponíveis do family_history_with_overweight: ", df['family_history_with_overweight'].unique())
```

#### Renomeando para o padrão e continuando com a troca de tipos

```python
df.rename(columns={'family_history_with_overweight': 'FHWO'}, inplace=True)
df["FHWO"] = df["FHWO"].replace({"no": 0, "yes": 1})
```

```python
print("Mostrando todos os valores disponíveis do SCC: ", df['SCC'].unique())
```

```python
df["SCC"] = df["SCC"].replace({"no": 0, "yes": 1})
```

```python
print("Mostrando todos os valores disponíveis do FAVC: ", df['FAVC'].unique())
```

```python
df["FAVC"] = df["FAVC"].replace({"no": 0, "yes": 1})
```

```python
print("Mostrando todos os valores disponíveis do SMOKE: ", df['SMOKE'].unique())
```

```python
df["SMOKE"] = df["SMOKE"].replace({"no": 0, "yes": 1})
```

```python
print("Mostrando todos os valores disponíveis do MTRANS: ", df['MTRANS'].unique())
```

```python
df["MTRANS"] = df["MTRANS"].replace(
    {
        "Public_Transportation": 0, 
        "Walking": 1, 
        "Automobile": 2, 
        "Motorbike": 3, 
        "Bike": 4,
    })
```

## Importando CSV

```python
df.to_csv('obesity_cleaned.csv', index=False)
```
